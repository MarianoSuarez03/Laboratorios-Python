## Análisis por Nivel Educacional Etnia de los Padres [1.5 Puntos]

El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos. 
Como científicos de datos, ud. y su equipo creen que deben encontrar evidencia para confirmar o refutar la hipótesis del rector.

Para esto, deciden generar dos análisis: una tabla de resumen por una parte y gráficos de caja por otro.

### Tabla de Resumen

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificación a través de un mapeo (investigar `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

### 1.3.4 Información por Categorías [1 punto]

````python
df_grades = df_grades.drop(["index"], axis = 1)

def alumnos_promedio(dataframe_in):
  dataframe_in['GPA'] = (dataframe_in['math score'] + dataframe_in['reading score'] + dataframe_in['writing score'])/3
  return dataframe_in

df_grades = alumnos_promedio(df_grades)
map_dict = {'some high school': 'school', 'some college': 'school', 'high school': 'school', "bachelor's degree": 'college', "associate's degree": 'college', "master's degree": 'postgraduate'}

def mapeo(dataframe_in):
    dataframe_in["simple parental level of education"] = dataframe_in["parental level of education"].map(map_dict)
    return dataframe_in

df_grades = mapeo(df_grades)

df_grades = df_grades.set_index(['race/ethnicity', 'simple parental level of education'])
df_avg = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).mean()
df_avg["count"] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size()
df_avg = df_avg.round(2)

def porcentaje(dataframe_in):
    total = dataframe_in["count"].sum()
    dataframe_in["porcentaje"] = (100*dataframe_in["count"]/total).round(2)
    #dataframe_in.astype({'porcentaje': 'str'})
    dataframe_in["porcentaje"] = dataframe_in["porcentaje"].astype(str) + " %"
    return dataframe_in

porcentaje(df_avg)
display(df_avg)

````

````python
other_grades =  pd.read_csv("/work/other_grades.csv").set_index('names')

df_grades1 = df_grades.set_index('names').join(other_grades, how='outer')
display(df_grades1)

df_grades2 = df_grades.set_index('names').join(other_grades, how='left')
display(df_grades2)

df_grades3 = df_grades.set_index('names').join(other_grades, how='right')
display(df_grades3)

df_grades4 = df_grades.set_index('names').join(other_grades, how='inner')
display(df_grades4)
````

> **Justificación:**



Como se puede apreciar dada la cantidad de columnas y el hecho de que no presentan datos faltantes, los mejores join son inner y left (utilizando el orden df_grades y other_grades). Lo anterior tiene sentido al entender como funciona cada uno:

* **inner join** corresponde a una intersección de los dataframes utilizando la llave de nombres, como df_grades contiene a los 875 alumnos, la intersección agrega las nuevas columnas solo para estos alumnos, obviando los demás registros.

* **left join** agrega la información al intersectar ambos dataframes, al dataframe de la izquierda, dejando como NaN los valores en donde el dataframe de la derecha no tiene valores para el primero. Como el primer dataframe corresponde al que tiene los 875, solo se agregan los valores de las nuevas columnas manteniendo el número de alumnos sin generar valores vacíos.

* **right join** no sirve en este caso, pues existen duplicados y una cantidad incorrecta de alumnos, por lo quedan datos vacíos debido a esto, manteniendo los problemas originales del dataset de la derecha.

* **outer join** nuevamente utiliza la información del dataset de la derecha, pues la tabla generada corresponde a la unión menos la intersección de los dataframes, como se puede apreciar, se mantienen las 1000 observaciones y hay presencia de valores vaciós en la tabla final.
